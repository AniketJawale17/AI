```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
type(iris)
```

```python
iris.target
```

```python
classifier=tree.DecisionTreeClassifier() #model
```

```python
classifier.fit(iris.data,iris.target)
```

```python
plt.figure(figsize=(15,10))
```

```python
tree.plot_tree(classifier)
```

```python
tree.plot_tree(classifier,filled=True)
```

```python
classifier2=tree.DecisionTreeClassifier(max_depth=2) #model for maximum depth as 2
```

```python
classifier2.fit(iris.data,iris.target)
```

```python
tree.plot_tree(classifier2)
```

```python
tree.plot_tree(classifier2,filled=True)
```
